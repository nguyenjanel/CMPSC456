# Stiff Differential Equations

Large rate of variation of solution and the consequentally small step-size required to obtain reasonable accuracy.

```python
# importing libraries
import numpy as np
import math
from matplotlib import pyplot as plt
from ipywidgets import interactive, fixed
```

```python
# --- Setup for Problem (a) ---
def dydt_a(t, y):
    return -9 * y

def fy_a(t, y):
    return -9

def actual_a(t):
    return math.exp(1 - 9*t)

# --- Setup for Problem (b) ---
def dydt_b(t, y):
    return -20*y + 20*np.sin(t) + np.cos(t)

def fy_b(t, y):
    return -20

def actual_b(t):
    return np.sin(t) + np.exp(-20*t)
```

```python
# Backward Euler's Method routine for solving Initial Value Problems (IVPs)
# Use with dydt to evaluate RHS of differential equation
# Input: interval inter, initial value y0, number of steps n
# FPI tolerance tol, and max number of FPI M
# Output: time steps t, approximate solution w
# Usage: backeuler(a,b,y_0,n,tol,M)
# Example: backeuler(0,1,1,10,5,20)
def trapezoid_newton(a, b, y0, n, tol, M):
    t = np.zeros(n+1)
    w = np.zeros(n+1)
    h = (b - a) / n
    w[0] = y0
    t[0] = a
    jmax = 1

    for i in range(1, n+1):
        t[i] = t[i-1] + h

        # Initial guess (Predictor): Explicit Euler
        w_old = w[i-1] + h * dydt(t[i-1], w[i-1])

        # Newton's Method loop
        for j in range(1, M+1):
            # f_now is f(t_i, w_i), f_next is f(t_{i+1}, w_old)
            f_now = dydt(t[i-1], w[i-1])
            f_next = dydt(t[i], w_old)

            # F(x) = x - w_i - (h/2)(f_now + f_next)
            F = w_old - w[i-1] - (h/2) * (f_now + f_next)

            # F'(x) = 1 - (h/2) * df/dy
            dF = 1 - (h/2) * fy(t[i], w_old)

            w_new = w_old - F / dF

            if abs(w_new - w_old) < tol:
                break
            w_old = w_new

        w[i] = w_new
        jmax = max(j, jmax)

    print(f"Convergence reached with jmax = {jmax}")
    return t, w

# Run and Print Table
t_vals, w_vals = trapezoid_newton(0, 2, 1, 8, 1e-6, 20)

print(f"{'t':>6} | {'w_i (Trap)':>12} | {'y(t)':>12} | {'Error':>12}")
for i in range(len(t_vals)):
    act = actual_b(t_vals[i])
    print(f"{t_vals[i]:6.2f} | {w_vals[i]:12.6f} | {act:12.6f} | {abs(act-w_vals[i]):12.6e}")
```
