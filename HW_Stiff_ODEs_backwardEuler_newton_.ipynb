# Stiff Differential Equations

Large rate of variation of solution and the consequentally small step-size required to obtain reasonable accuracy.

```python
# importing libraries
import numpy as np
import math
from matplotlib import pyplot as plt
from ipywidgets import interactive, fixed
```

```python
# --- Setup for Problem (a) ---
def dydt_a(t, y):
    return -9 * y

def fy_a(t, y):
    return -9

def actual_a(t):
    return math.exp(1 - 9*t)

# --- Setup for Problem (b) ---
def dydt_b(t, y):
    return -20*y + 20*np.sin(t) + np.cos(t)

def fy_b(t, y):
    return -20

def actual_b(t):
    return np.sin(t) + np.exp(-20*t)
```

```python
# Backward Euler's Method routine for solving Initial Value Problems (IVPs)
# Use with dydt to evaluate RHS of differential equation
# Input: interval inter, initial value y0, number of steps n
# FPI tolerance tol, and max number of FPI M
# Output: time steps t, approximate solution w
# Usage: backeuler(a,b,y_0,n,tol,M)
# Example: backeuler(0,1,1,10,5,20)
def backeuler(dydt, fy, a, b, y0, n, tol, M):
    # initialize/define
    t    = np.zeros(n+1) # independent var
    w    = np.zeros(n+1) # dependent var
    h    = (b - a) / n   # Step size calculation
    w[0] = y0            # Initial condition y(0) = 1
    t[0] = a             # Start time
    jmax = 1

    for i in range(1,n+1):
        t[i] = t[i-1] + h
        # Initial guess for use in Newton's Method
        w_old =  w[i-1] + h * dydt(t[i-1], w[i-1])

        # Iteration for Newton's Method
        # x_n+1 = x_n - F(x_n)/F'(x_n)
        for j in range(1,M+1):
            # F(x) = x - w_prev - h*f(t_curr, x)
            F = w_old - w[i-1] - h * dydt(t[i], w_old)

            # F'(x) = 1 - h*fy(t_curr, x)
            dF = 1 - h * fy(t[i], w_old)

            w_new = w_old - F / dF # Newton Update step

            # check if tolerance is met:
            if abs(w_new - w_old) < tol:
                w[i] = w_new
                break
            # tolerance not met, keep going if j.leq.m
            w_old = w_new

        jmax = max(j,jmax)

    print('jmax =',jmax)
    return t,w

def print_table(t, w, actual_func, title):
    print(f"\n--- {title} ---")
    print(f"{'t_i':>6} | {'w_i (BE)':>12} | {'y(t_i)':>12} | {'Error':>12}")
    print("-" * 50)
    for i in range(len(t)):
        act = actual_func(t[i])
        error = abs(act - w[i])
        print(f"{t[i]:6.2f} | {w[i]:12.6f} | {act:12.6f} | {error:12.6e}")

# --- Execution ---
# Problem (a)
t_a, w_a = backeuler(dydt_a, fy_a, 0, 1, math.e, 10, 1e-6, 20)
print_table(t_a, w_a, actual_a, "Problem (a): y' = -9y, h=0.1")

# Problem (b)
t_b, w_b = backeuler(dydt_b, fy_b, 0, 2, 1, 8, 1e-6, 20) # n=8 for h=0.25 on [0,2]
print_table(t_b, w_b, actual_b, "Problem (b): y' = -20y + 20sin(t) + cos(t), h=0.25")
```
